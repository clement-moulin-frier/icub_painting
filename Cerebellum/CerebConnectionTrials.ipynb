```python
def cs_us(x, y):
    us = float(norm([x, y]) > 1)
    cs = abs(1. - norm([x, y]))
    return cs, us
def cs_us_normal(x, y):
    us = float(norm([x, y])+randn()*0.01 > 1)
    cs = abs(1. - norm([x, y]))
    return cs, us
```
